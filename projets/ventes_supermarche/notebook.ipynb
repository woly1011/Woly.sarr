Ce notebook analyse les ventes contenues dans `dataset.csv`.

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("dataset.csv")
df["CA"] = df["Quantite"] * df["Prix_Unitaire"]

df.head()
```

```
# KPI â€” chiffre d'affaires total
df["CA"].sum()
```

```
# Top produits
top = df.groupby("Produit")["CA"].sum().sort_values(ascending=False)
top
```

```
# Visualisation
top.plot(kind="bar", figsize=(8,4))
plt.title("Top produits par chiffre d'affaires")
plt.show()
```
