# üåç Analyse des salaires dans le monde
Dataset : `dataset.csv`

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("dataset.csv")
df.head()
```

```
# Boxplot par pays
df.boxplot(column="Salaire", by="Pays", figsize=(8,5))
plt.title("Distribution des salaires par pays")
plt.show()
```

```
# Salaire moyen par pays
df.groupby("Pays")["Salaire"].mean()
```
